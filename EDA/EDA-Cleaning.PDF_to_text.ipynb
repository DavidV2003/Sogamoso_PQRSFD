```
import pandas as pd
import numpy as np
from pdfminer.high_level import extract_text #For PDF text extraction
from os import listdir #Extraction of files/folders names in a floder
from os.path import isfile, join #Extraction of only files (no folders)
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim 
import re
```

```
#Creation of an empty dataframe which will be populated with the readings of the PDFs
df_pdf = pd.DataFrame(columns = ['ID', 'DATE', 'SUBJECT','COUNTRY','ADDRESS','PQRSDF'])
```

We create functions that will read the data from the `PDF` files.

```
#Function to extract "radicado"
def ext_radicado(text_list):
    radicado = text_list[0].partition("N°. ")[2]
    if radicado == "":
        return np.nan
    else:
        return radicado 

#Function to extract "fecha_radicado"
def ext_fecha_radicado(text_list):
    fecha_radicado = text_list[1].partition("   Folios")[0]
    if fecha_radicado == "":
        return np.nan
    else:
        return fecha_radicado 

#Function to extract "asunto"
def ext_asunto(text_list):
    asunto = text_list[9].partition("Asunto : ")[2]
    if asunto == "":
        return np.nan
    else:
        return asunto

#Function to extract "pais"
def ext_pais(text_list):
    pais = text_list[-6]
    if pais == "":
        return np.nan
    else:
        return pais

#Function to extract "direccion"
def ext_direccion(text_list):
    direccion = text_list[-7]
    if direccion == "":
        return np.nan
    else:
        return direccion

#Function to extract "peticion"
def ext_peticion(text_list):
    peticion = text_list[10]
    lineas_peticion = [peticion]
    i=0
    while text_list[10+i] != text_list[-11]:
        i += 1
        lineas_peticion.append(text_list[10+i])
        peticion = ' '.join(lineas_peticion)
    return peticion
```

Now, we add the extracted data from 2018's `pdf` files to our dataframe `df_pdf`.

```
for pdf in listdir("/datasets/gdrive/DS4A - Team Project/data/PDF_Orfeo.v1/2018/200"):
    if isfile(join("/datasets/gdrive/DS4A - Team Project/data/PDF_Orfeo.v1/2018/200", pdf)):
        text = extract_text(join("/datasets/gdrive/DS4A - Team Project/data/PDF_Orfeo.v1/2018/200", pdf))
        text_list = text.split('\n')[:] = [x for x in text.split('\n') if x]
        df_pdf = df_pdf.append({ 
            'ID' : ext_radicado(text_list),
            'DATE' : ext_fecha_radicado(text_list),
            'SUBJECT' : ext_asunto(text_list),
            'COUNTRY' : ext_pais(text_list),
            'ADDRESS' : ext_direccion(text_list),
            'PQRSDF' : ext_peticion(text_list)
        },ignore_index=True )
```

## [DISCARDED] Georeferencing

Now, for georeferencing purposes, we associate to each address its (approx.) coordinates.

```
# WARNING: This code takes around 12min to compile
#Setting geocoding service (this one uses OSM)
geolocator = Nominatim(user_agent="facalderonm314@gmail.com")
#Setting delay time between request so the geocoding service provider does not block us
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
#Create new df and standarizing adresses
df_pdf_geo = df_pdf
df_pdf_geo['STD_ADDRESS'] = df_pdf_geo['ADDRESS'].apply(geocode)
```

```
# Create longitude, laatitude and altitude from location column (returns tuple)
df_pdf_geo['COORD'] = df_pdf_geo['STD_ADDRESS'].apply(lambda loc: tuple(loc.point) if loc else None)
# Split column into latitude, longitude and altitude columns
df_pdf_geo[['LAT', 'LONG','ALT']] = pd.DataFrame(df_pdf_geo['COORD'].tolist(), index=df_pdf_geo.index)
```
