# FE: Improving other datasets and extracting new data

## Improving the other datasets: aseguramiento.xlsx

```
aseguramiento = pd.read_excel('/datasets/gdrive/DS4A - Team Project/data/aseguramiento.v2/aseguramiento.xlsx', parse_dates=['fecha_entrada'])
aseguramiento.rename(columns = {'radicado':'radi_nume_radi'}, inplace = True)
aseguramiento
```

I create these functions to extract data from different datasets so we can add it to our clean df.

```
def get_municipio(muni_codi,dpto_codi):
    i = municipio.index[(municipio['muni_codi'] == muni_codi) & (municipio['dpto_codi'] == dpto_codi)].tolist()[0]
    return municipio[municipio["muni_codi"] == muni_codi].muni_nomb[i]
```

```
def get_departamento(dpto_codi):
    i = departamento.index[departamento['dpto_codi'] == dpto_codi].tolist()[0]
    return departamento[departamento["dpto_codi"] == dpto_codi].dpto_nomb[i]
```

```
def get_pais(id_pais):
    if id_pais == 170:
        return "COLOMBIA"
    if id_pais == 200:
        return "ESTADOS UNIDOS"
```

Now, let us enhance the `aseguramiento.xlsx` file.

```
# Merging "aseguramiento" with "direcciones" and correcting info
df1 = pd.merge(aseguramiento["radi_nume_radi"],direcciones[["radi_nume_radi","sgd_dir_direccion","muni_codi","dpto_codi","id_pais"]],on='radi_nume_radi',how='left')
# Changing the ID of each municipio for the actual name
df1["muni_nomb"] = df1.apply(lambda x: get_municipio(x.muni_codi, x.dpto_codi), axis=1)
df1.drop(["muni_codi"], axis=1,inplace=True)
# Changing the ID of each departamento for the actual name
df1["dpto_nomb"] = df1["dpto_codi"].apply(get_departamento)
df1.drop(["dpto_codi"], axis=1,inplace=True)
# Changing the ID of the country (Colombia = 170) for the actual name
df1["pais_nomb"] = df1["id_pais"].apply(get_pais)
df1.drop(["id_pais"], axis=1,inplace=True)
#df1
```

```
# Merging "aseguramiento" with "radicado" and correcting infoget_departamento
df2 = pd.merge(df1,radicado[["radi_nume_radi","radi_fech_radi","ra_asun"]],on='radi_nume_radi',how='left')
#df2
```

```
#final df
new_aseguramiento=df2[["radi_nume_radi","radi_fech_radi","ra_asun","sgd_dir_direccion","muni_nomb","dpto_nomb","pais_nomb"]]
new_aseguramiento['radi_fech_radi'] =  pd.to_datetime(new_aseguramiento['radi_fech_radi'])
new_aseguramiento.columns = ['RADICADO', 'FECHA', 'ASUNTO', 'DIRECCION', "MUNICIPIO", "DEPARTAMENTO", "PAIS"]
new_aseguramiento
```

Finally, we export the data.

```
new_aseguramiento.to_csv(r'/datasets/gdrive/DS4A - Team Project/data/aseguramiento.v3/aseguramiento_final.csv',encoding='UTF-8', index=False)
```
