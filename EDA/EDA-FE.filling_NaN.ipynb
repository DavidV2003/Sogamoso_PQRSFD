```
import pandas as pd
import numpy as np
import re
pd.options.mode.chained_assignment = None 
```

We import the unified dataset.

```
# Importing the df with all data
df=pd.read_csv('/datasets/gdrive/DS4A - Team Project/data/all_data.csv', encoding="UTF-8")
df.columns = [c.replace(' ', '_') for c in df.columns]
df
```

## Extracting EPABS

We start filtering the data to only have request with NaN values in the `EAPB` column and available text in the PQRSDF.

```
df_EAPB = df.loc[(df["EAPB"].isnull() == True) & (df["TEXTO_PQRSDF"].isnull() == False)]
df_EAPB
```

We use the EAPBs already present in the original dataset and this [Official Open Data](https://www.datos.gov.co/Salud-y-Protecci-n-Social/N-mero-de-afiliados-por-departamento-municipio-r-g/wctu-xgdj/data) to come up with a list of possible EAPBs mentioned in the PQRSDF's text.

```
df["EAPB"].unique()
```

```
# List of EAPB
EAPB_list =["COMFABOY", "COMFAMILIAR", "FERROCARRILES", 'SALUD TOTAL',
                "CAFESALUD", 'SANITAS', 'COMPENSAR', 'COOMEVA',
                'FAMISANAR', "CRUZ BLANCA", "SALUDVIDA",
                'NUEVA EPS', 'COMPARTA', 'MEDIMAS', "EMDISALUD",
                'MEDISALUD', 'COOSALUD', "ALIANSALUD", "SURAMERICANA",
                'ASMET SALUD', 'CAPRESOCA', 'ECOOPSOS',
                'CAJACOPI', "MUTUAL SER", 'CAPITAL SALUD', 
                'SUPERSALUD', 'MIGRANTE', 'SANIDAD',
                'CONVIDA', 'JER SALUD', 'EMMSANAR']

# List to str
EAPB_str = '|'.join(EAPB_list)
EAPB_str
```

And now we detect those entries where the PQRSDF's text has one of EAPB in the list above.

```
df_EAPB['EAPB_ENCONTRADA'] = df_EAPB.TEXTO_PQRSDF.str.findall(EAPB_str, flags = re.IGNORECASE)
df_EAPB["EAPB_ENCONTRADA"] = df_EAPB["EAPB_ENCONTRADA"].str[0].str.upper()
df_EAPB["EAPB_ENCONTRADA"].unique()
```

```
df_EAPB
```

```
df
```

We see that almost 23% of the entries had a match. It is not a big number, but nontheless is new data. Finally, we replace the old NaNs with this new info.

```
df = df.join(df_EAPB.EAPB_ENCONTRADA)
df["EAPB"] = df[["EAPB","EAPB_ENCONTRADA"]].fillna('').apply("".join,axis=1)
df.loc[df['EAPB']=='RETIRADO COOMEVA', 'EAPB'] = "COOMEVA"
df.loc[df['EAPB']=='', 'EAPB'] = np.nan
df.drop("EAPB_ENCONTRADA",axis=1,inplace=True)
df
```

## Extracting user gender

Using a similar technique as above, we search for "El usuario" and "La usuaria" in the PQRSDF's text to determine the user gender.

```
df_gen = df.loc[(df["GENERO"].isnull() == True) & (df["TEXTO_PQRSDF"].isnull() == False)]
df_gen
```

```
# List of wanted matches
gen_list =["USUARIO","USUARIA"]

# List to str
gen_str = '|'.join(gen_list)
gen_str
```

```
df_gen["GENERO_ENCONTRADO"] = df_gen.TEXTO_PQRSDF.str.findall(gen_str, flags = re.IGNORECASE)
df_gen["GENERO_ENCONTRADO"] = df_gen["GENERO_ENCONTRADO"].str[0].str.upper()
df_gen.loc[df_gen['GENERO_ENCONTRADO']=='USUARIO', 'GENERO_ENCONTRADO'] = "M" 
df_gen.loc[df_gen['GENERO_ENCONTRADO']=='USUARIA', 'GENERO_ENCONTRADO'] = "F" 
df_gen
```

We see that around 70% of the entries had a match. Finally, we replace the old NaNs with this new info.

```
df = df.join(df_gen.GENERO_ENCONTRADO)
df["GENERO"] = df[["GENERO","GENERO_ENCONTRADO"]].fillna('').apply("".join,axis=1)
df.loc[df['GENERO']=='', 'GENERO'] = np.nan
df.drop("GENERO_ENCONTRADO",axis=1,inplace=True)
df
```

Finally, we save this enhanced dataset.

```
df.to_csv(r'/datasets/gdrive/DS4A - Team Project/data/all_data_enhanced.csv',encoding='UTF-8', index=False)
```
